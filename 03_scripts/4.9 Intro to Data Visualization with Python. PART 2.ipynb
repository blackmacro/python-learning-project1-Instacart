# Extra. Checking forr correlation between all variables. Wondering if there are relations between original & derived variables. Also any type of correlation for any 'all vs all' variables. My thinking here is to find correlations in a format that is easy to read.

```python
df_insta.corr() # numerical expression
```

```python
corr=df_insta.corr()  #correlation in a chart format
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
